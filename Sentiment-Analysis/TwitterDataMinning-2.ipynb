```python
import pandas as pd 
import datetime
from tqdm.notebook import tqdm
import snscrape.modules.twitter as sntwitter
```

```python
#Getting the data before date
now = datetime.datetime.now()
one_day = datetime.timedelta(days=1)
one_day_before = now - one_day
formatted_time = one_day_before.strftime('%Y-%m-%d %H:%M')

profile = "elonmusk"

userScraper = sntwitter.TwitterProfileScraper(profile)
userdata = userScraper._get_entity()

print(f"Scraping {profile}")

tweets = []

for i, tweet in enumerate(userScraper.get_items()):
  if(tweet.date.strftime('%Y-%m-%d %H:%M')>formatted_time):
    data = [tweet.date,
            tweet.content,
            tweet.likeCount,
            tweet.retweetCount]

    tweets.append(data)

# for i, tweet in enumerate(userScraper.get_items()):
#   if(tweet.date.strftime('%Y-%m-%d %H:%M')>formatted_time):
#           print(f"#{i+1} of {userdata.statusesCount} on ({tweet.date.strftime('%Y-%m-%d %H:%M')}): {tweet.content}")
#           print(f"#{i+1} of{tweet.content}")
# print("Done!")
```

```python
tweets_df = pd.DataFrame(
    tweets,columns = ["Date","Content","LikeCount","ReTweetCount"]
    )

tweets_df
```
